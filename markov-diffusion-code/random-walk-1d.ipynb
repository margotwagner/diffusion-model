```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1-D random walk for one molecule from a range of x = [-10, 10]

# Equal probability to move left or right
p = 0.5

# Initialize starting position (-10 to 10 range)
start = 0 
positions = [start]

# creating random points
r = np.random.random(1000)

for left_i, right_i in zip(r < p, r >= p):
    left = left_i and positions[-1] > -10
    right = right_i and positions[-1] < 10
    positions.append(positions[-1] - left + right)

# plotting down the graph of the random walk in 1D
plt.plot(positions)
plt.show()
```

```python
# 1-D random walk for 3 molecules from a range of x = [-10, 10]

# Equal probability to move left or right
p = 0.5

# number of molecules
N = 3

for n in range(N):
    # Initialize starting position (-10 to 10 range)
    start = 0 
    positions = [start]

    # creating random points
    r = np.random.random(1000)

    for left_i, right_i in zip(r < p, r >= p):
        left = left_i and positions[-1] > -10
        right = right_i and positions[-1] < 10
        positions.append(positions[-1] - left + right)

    # plotting down the graph of the random walk in 1D
    plt.plot(positions)

plt.show()
```
